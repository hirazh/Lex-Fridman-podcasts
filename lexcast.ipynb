## Lex Fridman Podcast

Obtainig data from lex Fridman's podcast from YouTube playlist

Available Data untill 12 October 2022

```python
from pytube import YouTube
from pytube import Playlist
```

```python
p = Playlist("https://www.youtube.com/watch?v=iZRbD7q1n-U&list=PLrAXtmErZgOdP_8GztsuKi9nrraNbKKp4")

playlist_result = []

for v in p.videos:
    # print(v.title)
    dic_temp = {"Title":"", "Date": "", "Length": "", "Views":""}

    dic_temp["Title"] = v.title
    dic_temp["Date"] = v.publish_date
    dic_temp["Length"] = v.length
    dic_temp["Views"] = v.views

    playlist_result.append(dic_temp)
```

```python
playlist_result
```

```python
import pandas as pd

lexcast_rawdata = pd.DataFrame.from_dict(playlist_result)
lexcast_rawdata
```
